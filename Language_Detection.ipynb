```python
# the flow chart  of this project will be  ----     language ->  Numbers -> Machine Learning Model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer   # transform the text data into numerical data
# it also collects unique words which is called bag of words
```

```python
# we will use Naive Bayes Algorithm in this
from sklearn.naive_bayes import MultinomialNB

# mostly used for word counts / text classification
```

```python
data['language'].value_counts()
```

```python
# Convert  it into array

x = np.array(data['Text'])
y = np.array(data['language'])
```

```python
print(x)
```

```python
print(y)
```

```python
# we are mostly concern with the text column  , as bag of words concept will apply there only , so the next step
# is be to convert the array into numbers to the text column
```

```python
cv = CountVectorizer()
```

```python
X = cv.fit_transform(x)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.33, random_state =42)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
# Model Building

model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
# Model Evalution

model.score(X_test,y_test)
```

```python
user = input("Enter a Text :  ")
data = cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```
