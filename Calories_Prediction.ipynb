```python
Both.replace({'Gender':{'male':0 , 'female':1}} , inplace = True)
```

```python
Both.head()
```

```python
x = Both.drop(columns = ['Calories'  , 'User_ID'], axis =1)
y = Both['Calories']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2) # Corrected the variable name from y_train to y_test
```

```python
xgb = XGBRegressor()
```

```python
xgb.fit(x_train ,y_train)
```

```python
prediction = xgb.predict(x_test)
```

```python
# it tells us that how much error is our model making

accuracy  = metrics.mean_absolute_error(y_test , prediction)
```

```python
print(accuracy)
```
